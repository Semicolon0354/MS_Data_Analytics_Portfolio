# import necessary packages 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ast import literal_eval

#load csv as dataframe
Resultsdf=pd.read_csv('TSP_Results.csv')

#fix format inconsistency with genetic route
Resultsdf['Best Route']=Resultsdf['Best Route'].apply(literal_eval)
for i in range(len(Resultsdf)):
    if Resultsdf['Algorithm'][i]=='genetic':
        Resultsdf['Best Route'][i].insert(0,0)
#calculate route length for comparison and sorting
Resultsdf['Route Length']= Resultsdf['Best Route'].apply(lambda x: len(x))

#drop the exhaustive test row and Best Route Column for cleaner visualization
Resultsdf=Resultsdf.drop(index=0)
Resultsdf=Resultsdf.reset_index(drop=True)
Resultsdf=Resultsdf.drop(columns='Best Route')

#group the results by route length (and data set) then sort by distance and display the results
bestdistancedf=Resultsdf.groupby('Route Length').apply(lambda x: x.sort_values(by='Best Distance'))
bestdistancedf

```python
#load post-optimization results as dataframe
Results2df=pd.read_csv('TSP_Results_updated.csv')

#calculate route length (number of cities) and drop Best Route column
Results2df['Best Route']=Results2df['Best Route'].apply(literal_eval)
Results2df['Route Length']= Results2df['Best Route'].apply(lambda x: len(x))
Results2df=Results2df.drop(columns='Best Route')

#add genetic results from first run for comparison
for i in range(len(Resultsdf)):
    if Resultsdf['Algorithm'][i]=='genetic':
        Results2df.loc[len(Results2df)+1]=Resultsdf.iloc[i]
        
#apply same group and sort as the other dataset and display results 
bestdistance2df=Results2df.groupby('Route Length').apply(lambda x: x.sort_values(by='Best Distance'))
bestdistance2df
```
